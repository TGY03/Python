```python
from ultralytics import YOLO

# Load a model
model = YOLO("./yolov8n.pt")  # load a pretrained model (recommended for training)

# 預測
results = model("./party.jpg")

# 顯示物件類別
print(results[0].boxes.cls)
print()

# 顯示物件座標
print(results[0].boxes.xyxy)
```
